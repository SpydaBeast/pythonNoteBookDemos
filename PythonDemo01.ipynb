# Activity 9: Pong Game Development

```python
import turtle
win = turtle.Screen()
win.title("Game creating")
win.setup(width=900, height=500)
win.tracer(0)

# Game Objects

# Player 1
player_1 = turtle.Turtle()
player_1.speed(0)
player_1.shape("square")
player_1.color("black")
player_1.shapesize(stretch_wid=5, stretch_len=1)              
player_1.penup()
player_1.goto(-350, 0)

# Player 2
player_2 = turtle.Turtle()
player_2.speed(0)
player_2.shape("square")
player_2.color("black")
player_2.shapesize(stretch_wid=5, stretch_len=1)              
player_2.penup()
player_2.goto(350, 0)

# ball
ball = turtle.Turtle()
ball.speed(0)
ball.shape("square")
ball.color("black")              
ball.penup()
ball.goto(0, 0)

#Ball speed
ball.dx = .2
ball.dy = .2

#Score Board
score = turtle.Turtle()
score.speed(0)
score.color("black")
score.penup()
score.hideturtle()
score.goto(0, 200)
score.write("Player 1: 0 Player 2: 0", align="center", font=("Courier", 24, "normal"))

score_1 = 0
score_2 = 0

# Move player 1 paddle Function
def paddle_1_up():
    y = player_1.ycor()
    y += 20
    player_1.sety(y)

def paddle_1_down():
    y = player_1.ycor()
    y -= 20
    player_1.sety(y)    
    
# Move player 2 paddle Function
def paddle_2_up():
    y = player_2.ycor()
    y += 20
    player_2.sety(y)

def paddle_2_down():
    y = player_2.ycor()
    y -= 20
    player_2.sety(y)
    

    
# Key binding
win.listen()
win.onkeypress(paddle_1_up, "w")
win.onkeypress(paddle_1_down, "s")
win.onkeypress(paddle_2_up, "Up")
win.onkeypress(paddle_2_down, "Down")

#Moving the Ball
# main game loop
while True:
    win.update()
    
    #Move ball
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    # Check borders
    if ball.ycor() > 230:
        ball.sety(230)
        ball.dy *= -1
    
    if ball.ycor() < -230:
        ball.sety(-230)
        ball.dy *= -1
        
    if ball.xcor() > 430:
        ball.goto(0, 0)
        ball.dx *= -1
        score_1 += 1
        score.clear()
        score.write("Player 1: {} Player 2: {}".format(score_1, score_2) , align="center", font=("Courier", 24, "normal"))

        
    if ball.xcor() < -430:
        ball.goto(0, 0)
        ball.dx *= -1
        score_2 += 1
        score.clear()
        score.write("Player 1: {} Player 2: {}".format(score_1, score_2) , align="center", font=("Courier", 24, "normal"))

# Paddle Strike
    if ball.xcor() > 340 and ball.xcor() < 350 and (ball.ycor() < player_2.ycor() + 40 and ball.ycor() > player_2.ycor() -40):
        ball.setx(340)
        ball.dx *= -1
        
    if ball.xcor() < -340 and ball.xcor() < -350 and (ball.ycor() < player_1.ycor() + 40 and ball.ycor() > player_1.ycor() -40):
        ball.setx(-340)
        ball.dx *= -1
        

```
